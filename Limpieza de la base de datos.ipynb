```python
# Importamos las librerías necesarias 
import pandas as pd
```

```python
# Importamos el archivo
# Importamos los datos
df = pd.read_csv('datos_cat.csv')
df.head()
```

```python
# Revisamos los valores únicos para cada columna
print('Cole_depto_ubicacion:',df['cole_depto_ubicacion'].unique())
print('----------------------------------------------------------')
print('fami_estratovivienda:',df['fami_estratovivienda'].unique())
print('----------------------------------------------------------')
print('fami_personashogar:',df['fami_personashogar'].unique())
print('----------------------------------------------------------')
print('fami_tieneautomovil:',df['fami_tieneautomovil'].unique())
print('----------------------------------------------------------')
print('fami_tienecomputador:',df['fami_tienecomputador'].unique())
print('----------------------------------------------------------')
print('fami_tieneinternet:',df['fami_tieneinternet'].unique())
print('----------------------------------------------------------')
print('fami_tienelavadora:',df['fami_tienelavadora'].unique())
```

```python
# Eliminamos los que no corresponden
df = df[df['fami_estratovivienda'] != 'Técnica o tecnológica completa']
df = df[df['fami_personashogar'] != 'Estrato 4']
df = df[df['fami_tieneautomovil'] != '3 a 4']
```

```python
# Revisamos que se haya eliminado bien
print('Cole_depto_ubicacion:',df['cole_depto_ubicacion'].unique())
print('----------------------------------------------------------')
print('fami_estratovivienda:',df['fami_estratovivienda'].unique())
print('----------------------------------------------------------')
print('fami_personashogar:',df['fami_personashogar'].unique())
print('----------------------------------------------------------')
print('fami_tieneautomovil:',df['fami_tieneautomovil'].unique())
print('----------------------------------------------------------')
print('fami_tienecomputador:',df['fami_tienecomputador'].unique())
print('----------------------------------------------------------')
print('fami_tieneinternet:',df['fami_tieneinternet'].unique())
print('----------------------------------------------------------')
print('fami_tienelavadora:',df['fami_tienelavadora'].unique())
```

```python
# Descargamos el archivo limpio
df.to_csv('datos_limpio.csv', index=False)
```

```python
# Creamos como un .sql
# Nombre de la tabla
table_name = "tabla_data_final"

# Generar el script SQL manualmente para data_mod
script = "CREATE TABLE {} (\n".format(table_name)

for col in df.columns:
    script += "    {} VARCHAR(255),\n".format(col.replace('\'', ''))

script = script.rstrip(",\n") + "\n);\n\n"

script += "COPY {} (".format(table_name)

for col in df.columns:
    script += "{}, ".format(col)
script = script.rstrip(", ")
script += ") FROM stdin;\n"

for i in range(df.shape[0]):
    row = "\t".join([str(x) for x in df.iloc[i]])  # Crear una fila como una cadena de texto separada por tabulaciones
    script += "{}\n".format(row)

script += "\\."

# Guardar el script SQL en un archivo
with open('data_proy3.sql', 'w') as file:
    file.write(script)
```
